```python
import pandas as pd
```

```python
df=pd.read_csv('기아_고객_데이터_완성.csv')
```

```python
df.head()
```

```python
df['구매한제품'].unique()
```

```python
# 1. True가 하나라도 있는 제품 목록 구하기
eco_friendly_products = df.groupby('구매한제품')['친환경차'].any()

# 2. 그 중 True인 제품만 필터링 (index가 제품 이름)
eco_friendly_products = eco_friendly_products[eco_friendly_products == True].index

# 3. 원본 df에서 해당 제품만 필터링
df_filtered = df[df['구매한제품'].isin(eco_friendly_products)]
```

```python
df_filtered['구매한제품'].unique()
```
